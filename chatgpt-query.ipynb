## Инициализация OpenAI

1. Укажите свой OpenAI API-ключ.

2. Выберите модель, которую хотите использовать. Из списка https://platform.openai.com/docs/models/model-endpoint-compatibility (строка /v1/chat/completions).

3. Опционально: укажите параметры из списка: https://platform.openai.com/docs/api-reference/chat/create.

```python
import openai

openai.api_key = "sk-...A9ED"

PARAMS = {
    "model": "gpt-4",
    # add optional parameters here
}

def ask(system_prompt, user_prompt):
    messages = [
        {
            "role": "system",
            "content": system_prompt
        },
        {
            "role": "user",
            "content": user_prompt
        }
    ]

    completion = openai.ChatCompletion.create(messages=messages, **PARAMS)
    return completion.choices[0]["message"]["content"]
```
